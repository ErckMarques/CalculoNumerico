# Estudo gráfico

```python
import numpy as np
import math
import matplotlib.pyplot as plt

# cria uma sequência de n números uniformemente espaçados entre os limites (a,b) dados: np.linspace(a,b,n)
x = np.linspace(-5,5,10)

def g(x):
  return 2**(-x)
plt.plot(x, g(x), color = 'blue') #azul

hx = []
for i in range(len(x)):
    hx.append(2*math.sin(x[i]))
    i += 1
plt.plot(x, hx, color = 'red') #vermelho

f = g(x) - hx
plt.plot(x, f, color = 'orange') #laranja

plt.grid() # facilita a visualização
plt.xticks(np.arange(-5, 6, step=1)) # passo do eixo x
plt.show()
```

**Considerando:**

*   Gráfico
*   Teorema de Bolzano
*   O fato de f(x) ser sempre decrescente no intervalo [a ; b]

**Posso concluir que:**

I = [0 ; 1] é intervalo de separação da função f definida. Nesse intervalo a função possui apenas 1 raiz real.

# Método da falsa posição (cordas)

```python
import numpy as np
import math
import matplotlib.pyplot as plt

# cria uma sequência de n números uniformemente espaçados entre os limites (a,b) dados: np.linspace(a,b,n)
x = np.linspace(-1,2)
def g(x):
  return 2**(-x)

hx = []
for i in range(len(x)):
    hx.append(2*math.sin(x[i]))
    i += 1

f = g(x) - hx
plt.plot(x, f, color = 'black')


def f(x):
  return 2**(-x) - 2*math.sin(x)


# A quebra é realizada no ponto de interseção da reta (a,f(a)) e (b,f(b)) com o eixo x.
a = 0; b = 1;
#traçando a reta
plt.plot([a,b],[f(a),f(b)],color = 'red')

plt.grid() # facilita a visualização
plt.xticks(np.arange(-1, 3, step=1)) # passo do eixo x
plt.show()
```

```python
import math
def cordas(a,b,l):
  c = math.fabs(b - a) #distancia do intervalo
  i = 0
  if f(a)*f(b) < 0: # ajuste do intervalo
    while (c > l):

      #c = math.fabs(b - a)
      xi = (a*f(b)-b*f(a))/(f(b)-f(a))
      #print(i, "%.6f"%a,"%.6f"%b,"%.6f"%f(a),"%.6f"%f(b),"%.6f"%xi,"%.6f"%f(xi),"%.6f"%c)

      if f(xi) == 0:
        return xi, c
      else:
        return None, None

      if f(a)*f(xi) < 0:
        b = xi
      else:
        a = xi

      c = math.fabs(b-a)
      i += 1

    return xi, c
  else:
    print("Não há raiz neste intervalo")
```

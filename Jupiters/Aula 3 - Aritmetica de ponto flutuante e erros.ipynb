# Conversão de decimal para binário

## números inteiros

```python
numero = input("Digite um número para ser convertido em binário: ")

quociente = float(numero)
binario = []

while quociente > 0:
  resto = quociente % 2 #resto da divisão
  binario.append(resto) #digitos do numero binario
  quociente = quociente // 2 #divisão inteira

print('O número',numero, 'em binário é escrito como:', binario[::-1]) #imprime o inverso da minha lista para retornar meu número binário
```

## números com parte fracionária

```python
numero = float(input("Digite um número para ser convertido em binário: "))

inteiro, decimal = str(numero).split(".")
binario = []

while (numero < 2) and (int(decimal) != 0): #realiza as operações até o produto não ter parte fracionária
  numero = (int(decimal)/(10**len(decimal))) * 2
  inteiro, decimal = str(numero).split(".") #pega a parte inteira do numero
  binario.append(int(inteiro)) #digitos do numero binario

print(binario)
```

Representação de erros

```python
def calculaErros(valor_exato,valor_aproximado):
  erro_absoluto = valor_exato - valor_aproximado
  print('Erro absoluto:',erro_absoluto)

  erro_relativo = erro_absoluto/valor_exato
  print('Erro relativo:',erro_relativo)

  erro_percentual = 100 * erro_relativo
  print('Erro percentual:',erro_percentual)
```

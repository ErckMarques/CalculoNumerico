# Funções

```python
def classificarIdade(idade):
  if idade >= 65:
    print('Essa pessoa é idosa e tem', idade, 'anos')
  elif idade > 18:
    print('Essa pessoa é maior de idade, mas não é idosa, pois tem', idade, 'anos')
  else:
    print('Essa pessoa é menor de idade e tem', idade, 'anos')

#classificarIdade(45)
```

# Estruturas de repetição - laços/loops

# Matrizes

```python
# é uma lista de listas

m = [[1,2,3],[4,5,6],[7,8,9]]
m
```

```python
# função para percorrer a matriz
def lerMatriz(matriz):
  linhas, colunas = (len(matriz), len(matriz[0]))
  print(linhas, colunas)
  for i in range(linhas):
    for j in range(colunas):
      print(matriz[i][j])

lerMatriz(m)
```

```python
numeros = list(range(1,11))
numeros
```

```python
soma = 0
for i in range(len(numeros)):
  soma = soma + numeros[i]
print(soma)
```

```python
teste_de_soma = sum(numeros)
print(teste_de_soma)
```

```python
import math
print(math.fsum(numeros))
```

## INICIALIZAÇÃO DE MATRIZES

```python
import numpy as np #biblioteca para manipulacao de dados
mA = np.zeros((3,5))
print(mA)
```

```python
mB = np.ones((3,5))
print(mB)
```

```python
# matriz identidade
I = np.eye((5))
print(I)
```

```python
# matriz diagonal
mC = np.diag([1,2,3,4])
print(mC)
```

```python
# abaixo ou acima da diagonal principal
mD = np.diag([1,1,1,1],-1)
print(mD)
print('\n')
mE = np.diag([1,1,1,1],+1)
print(mE)
```

```python
# matriz aleatória
R = np.random.rand(3,5)
print(R)
```

```python
# matriz a partir de listas
L = [[1,2,3,4],[5,6,7,8]]
L = np.array(L) #converter a lista em um array
print(L)
```

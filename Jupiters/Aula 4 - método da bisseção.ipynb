###Estudo gráfico

```python
import numpy as np
import math
import matplotlib.pyplot as plt

# cria uma sequência de n números uniformemente espaçados entre os limites (a,b) dados: np.linspace(a,b,n)
x = np.linspace(-5,5,10)

def g(x):
  return 2**(-x)
plt.plot(x, g(x), color = 'blue') #azul

hx = []
for i in range(len(x)):
    hx.append(2*math.sin(x[i]))
    i += 1
plt.plot(x, hx, color = 'red') #vermelho

f = g(x) - hx
plt.plot(x, f, color = 'orange') #laranja

plt.grid() # facilita a visualização
plt.xticks(np.arange(-5, 6, step=1)) # passo do eixo x
plt.show()
```

```python
# fazendo um "zoom" na função f(x) para visualizar melhor seu comportamento no intervalo
f_cortado = g(x)[4:7] - hx[4:7]
plt.plot(x[4:7], f_cortado, color = 'black')
plt.grid()
plt.xticks(np.arange(-1, 3, step=1))
plt.show()

# todas as funções juntas
plt.plot(x[4:7], g(x)[4:7], color = 'blue')
plt.plot(x[4:7], hx[4:7], color = 'red')
plt.plot(x[4:7], f_cortado, color = 'orange')
plt.grid()
plt.xticks(np.arange(-1, 3, step=1))
plt.show()
```

###Estudo analítico

```python
import math

def f(x):
  return 2**(-x) - 2*math.sin(x)

def verificaRaiz(a,b):

  # em x = a:
  print(f(a))
  # em x = b:
  print(f(b))

  #Teorema de Bolzano:
  if (f(a)*f(b)) < 0:
    print("De acordo com o teorema de Bolzano, existe pelo menos uma raiz real de f no intervalo considerado.")

verificaRaiz(0,1)
```

**Considerando:**

*   Gráfico
*   Teorema de Bolzano
*   O fato de f(x) ser sempre decrescente no intervalo [a ; b]

**Posso concluir que:**

I = [0 ; 1] é intervalo de separação da função f definida. Nesse intervalo a função possui apenas 1 raiz real.

### Algoritmo da Bisseção

1) Defina a função que está trabalhando

2) Defina quem são a, b (intervalo de separação) e l (amplitude final)

3) Calcula a amplitude de [a;b]: c

4) Calcula o ponto médio de [a;b]: x0

5) Iterativamente avalia em qual subintervalo a raiz está:
a avaliação se repete enquanto (c > l) **ou** (f(x0) != 0)

6) Depois que algum dos critérios de parada foi atingido, escreva: O valor aproximado da raiz procurada é "x0". Em seguida, finalize o programa.

```python
b = np.random.rand(3,1)
print(b)
```

**2) Qual o valor do primeiro multiplicador? Não utilize os números e sim referências a matriz.**

```python
pivo = A[0,0]
m = -(A[1,0]/pivo)

linha0 = A[[0],:]
linha1 = A[[1],:]

cA = A #copia de A
cA[1] = (m*linha0)+linha1
print("A = ",cA)
print()
cb = b #copia de b
cb[1] = (m*cb[0])+cb[1]
print("b = ",cb)
```

**3. Determine o próximo multiplicador e atualize a linha 2 das matrizes A e b.**

```python
m = -(A[2,0]/pivo)

linha2 = A[[2],:]
cA[2] = (m*linha0)+linha2
print("A = ", cA)
print()
cb[2] = (m*cb[0])+cb[2]
print("b = ",cb)
```

**4. O trabalho com a primeira coluna está finalizado. Determine o próximo pivô.**

```python
pivo = A[1,1]
```

**5. Determine o próximo multiplicador e atualize a linha 2.**

```python
m = -(A[2,1]/pivo)

linha1 = A[[1],:]
linha2 = A[[2],:]
cA[2] = (m*linha1)+linha2
print("A = ",cA)
print()
cb[2] = (m*cb[1]) + cb[2]
print("b = ",cb)
```

# Eliminação Gaussiana sem pivoteamento

**Função que realiza a primeira parte da eliminação gaussiana, convertendo a matriz em uma matriz triangular superior.**

```python
import numpy as np
```

```python
matriz_coeficientes = np.array([[3,2,7],[-6,5,1],[3,11,-1]])
print(matriz_coeficientes)
print()
termos_independentes = np.array([8,-10,-9])
print(termos_independentes)
```

```python
def eliminacao_gaussiana(matriz_coeficientes, termos_independentes):
    '''
    matriz_coeficientes é uma matriz n x n
    termos_independentes é uma matriz n por 1
    '''

    A = matriz_coeficientes
    b = termos_independentes
    tamanho = len(A)

    for k in range(0, tamanho-1):
        pivo = A[k,k]
        for i in range(k+1, tamanho):
            m = - (A[i,k] / pivo)
            A[i] = (m * A[k]) + A[i]
            b[i] = (m * b[k]) + b[i]

    return A, b
```

```python
A,b = eliminacao_gaussiana(matriz_coeficientes, termos_independentes)
print(A)
print()
print(b)
```

**Função que determina as soluções x[j] de um sistema com matriz de coeficientes triangular superior.**

```python
def triangular_superior(A, b):
    '''
    A: matriz triangular superior
    b: termos independentes do sistema triangular superior

    retorna a solução do sistema
    '''
    tamanho = len(A)

    x = np.zeros(tamanho)
    x[tamanho-1] = b[tamanho-1] / A[tamanho-1, tamanho-1] # ultimo x[j]

    #range(inicio,parada,passo)
    for k in range(tamanho-2,-1,-1):
        for t in range(k,tamanho):
            b[k] = b[k] - (A[k,t] * x[t])

        x[k] = b[k] / A[k,k]

    return x
```

```python
x = triangular_superior(A, b)
print(x)
```

```python
print(matriz_coeficientes @ x)
```

# E se um pivô for igual a zero?

```python
def resolver_sistema(matriz_coeficientes, termos_independentes):
    tA, tb = eliminacao_gaussiana(matriz_coeficientes, termos_independentes)
    x = triangular_superior(tA, tb)
    return x
```

```python
matriz_coeficientes[0,0] = 10 ** -16
print(matriz_coeficientes)
print()

x = resolver_sistema(matriz_coeficientes,termos_independentes)
print('x = ',x)
```

```python
matriz_coeficientes = np.array([[3,2,1,-1],[0,0,1,3],[0,-3,-5,7],[0,2,4,0]])
termos_independentes = np.array([5,6,7,15])
print(matriz_coeficientes)
print()
x = resolver_sistema(matriz_coeficientes,termos_independentes)
print('x = ',x)
```
